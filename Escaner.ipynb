```python
import cv2 
import numpy as np 
from pyzbar.pyzbar import decode
import pywhatkit as kit
from datetime import datetime
import time
import pyautogui as pg
import pandas as pd
import sqlite3

cap = cv2.VideoCapture(0)

while True:
    success, img = cap.read()
    
    if not success:
        break
    
    for code in decode(img):
        
        decode_informacion = code.data.decode("utf-8")
        
        print(decode_informacion)
        
        rect_pts = code.rect
        
        if decode_informacion:
            pts = np.array([code.polygon], np.int32)
            cv2.polylines(img, [pts], True, (255, 0, 0), 3)
            cv2.putText(img, str(decode_informacion), (rect_pts[0], rect_pts[1]), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 255, 0),)

        cap.release()

        cv2.imshow("QR Code Content", img)
        cv2.waitKey(0)

decode_informacion = int(decode_informacion)

cv2.destroyAllWindows()

archExcel = 'datos_venta.xlsx'
df = pd.read_excel(archExcel, engine='openpyxl')

conn = sqlite3.connect('EstudiantesBD.db')

df.to_sql('Alumnos', conn, if_exists='replace', index=False)

fila_seleccionada = df.loc[df['folio'] == decode_informacion]

if not fila_seleccionada.empty:
    nTelefono = fila_seleccionada['celular'].values[0]
    print(f"- \n{fila_seleccionada}")
    print(f"-{nTelefono}")
else:
    print("No se encontraron resultados")

conn.close()

nTelefono = str(nTelefono)

tiempo = datetime.now()

fecActual = tiempo.strftime("%Y-%m-%d")
hActual = tiempo.strftime("%H:%M")
minActual = tiempo.strftime("%M")

num_tel = str("+52" + nTelefono)
num_tel

message = "El dia " + fecActual + "acaba de llegar tu hijo a las " + hActual
second = time.time() + 60
date = datetime.fromtimestamp(second)

kit.sendwhatmsg(num_tel,message, date.hour, date.minute)
pg.hotkey('enter')
```
